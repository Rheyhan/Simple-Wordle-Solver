# Basic algorithm. This will be refubrished later.

```python
import re
import json
```

```python
with open('simple_english_dictionary.json', 'r') as f:
    dictionary = json.load(f)

content = [i for i in dictionary.keys() if len(i) == 5]
content = [i for i in content if re.match("^[A-Za-z]{5}$", i)]
content = [i.lower() for i in content]
```

```python
# Shit baseline algorithm

letter1 = '.'
letter2 = 'p'
letter3 = 'i'
letter4 = 'n'
letter5 = 'e'

search = (letter1 + letter2 + letter3 + letter4 + letter5).lower()
blacklisted = 'abdcrms'.lower()
possible = ''.lower()


pattern = search.replace('.', '[a-z]')

filtered = [word for word in content if re.match(f'^{pattern}$', word)]
if blacklisted:
    filtered = [word for word in filtered if not any(char in word for char in blacklisted)]
filtered = sorted([word for word in filtered if all(char in word for char in possible)])

# All possible words
print(filtered)
```
